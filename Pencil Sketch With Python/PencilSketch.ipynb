Image to Pencil Sketch with Python



1. We need to read the image in RBG format 
2. and then convert it to a grayscale image.This will turn an image into a classic black and white photo.
3. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image.
4. Inversion can be used to enhance details. 
5. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. 
6. This can be done by dividing the grayscale image by the inverted blurry image.
Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
#Read the image
img=cv2.imread("/pencil_sketch/pexels-flickr-151511.jpg")
```

```python
img = cv2. resize(img, (400, 400))
cv2_imshow(img)
```

```python
gray_image=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
#invert the image
inverted_gray_image=255 - gray_image
cv2_imshow(inverted_gray_image)
```

```python
#blur the image by Gaussian function
blurred_image =cv2.GaussianBlur(inverted_gray_image,(21,21),0)
cv2_imshow(blurred_image)
```

```python
#invert the blur image
inverted_blur_img=255-blurred_image
cv2_imshow(inverted_blur_img)
```

```python
#pencil sketch
penciil_img=cv2.divide(gray_image,inverted_blur_img,scale=256.0)
cv2_imshow(penciil_img)
```
